```python
df['rating'].value_counts()
```

** ESRB encoding decryption **


* "E" ("Everyone") - "For all" *: Games with this rating may contain minimal violence, mostly "cartoon" character. Initially "k-a". * Replace "K," on "E" *

* EC - Early Childhood *: for preschool children. * Can be replaced by "E" *

* "E10 +" ("EVERYONE 10 AND OLDER") - "For all from 10 years and older" *: Projects with this rating may contain or slightly more multiplication or soft violence, or several candid scenes, or minimal blood. The rating was accepted by ESRB on March 2, 2005.

* "T" ("Teen") - "Teenagers" *: The game is suitable for persons from 13 years. Projects from this category may contain violence, obscene scenes, rough humor, in moderation of frank sexual content, blood or infrequent use of abnormative vocabulary.

* "M" ("Mature") - "For adults" *: The materials of the game are not suitable for adolescents under 17 years of age. Projects with this rating may contain enough ill violence, a large amount of blood with dismemberment, obscene sexy scenes or gross abnormative vocabulary, unwanted for younger audience.

* "AO" ("adults only 18+") - "Adult only" *: The content of the game is only for adults over 18 years old. Products from this category may contain long-term scenes of cruel violence and / or very frank sexual contents, as well as scenes with exposure. * Can be replaced by "M" *

* "RP" ("Rating Pending") - "Ranking is expected" *: The product was sent to ESRB and expects a rating assignment. This logo is used only on promotional presentations and in demo versions of games to the official date of sale on sale. * Replace on Nan *

Replace the rating values on the principle described above.

```python
df['rating'] = df['rating'].replace({'EC': 'E', 'AO': 'M', 'RP': np.nan,'K-A':'E'})
df['rating'].value_counts()
```

Check the amount of empty cells and try to fill them, if possible.

```python
df['rating'].isna().sum()
```

Fill in the rating values from other platforms.

```python
# If there is a choice of several rating values, we will accidentally take the first string.
df.rating.fillna(df.groupby('name')['rating'].transform('first'), inplace=True)
```

ESRB as an organization was founded in 1994 and assigned ratings for games manufactured in the United States and Canada.Accordingly, I assume that no meanings are missing for games common in other markets or issued until 2000, when the rating assignment practice has become massive.Check my hypothesis.

```python
df.groupby('year_of_release')['rating'].agg(
    lambda x: x.isna().sum()/len(x)).plot()
plt.title('Percentage of games without rating by year', fontsize=14)
plt.show()
```

As expected, most of the games without the rating were released until 2000.Now let's try to determine for which region there is not the lack of rating values.I assume that since this rating was introduced for the US and Canada markets, the lack of importance will be characteristic of other markets.

We will proceed from logic that 50% of profits will be in the home area of the game.

```python
# Wake a feature for calculating the game region
def region_game(row):
    na_sales = row['na_sales']
    eu_sales = row['eu_sales']
    jp_sales = row['jp_sales']
    other_sales = row['other_sales']
    total = row['total_sales']
    if (na_sales/total) > 0.5:
        return 'North_America'
    elif (eu_sales/total) > 0.5:
        return 'Europe'
    elif (jp_sales/total) > 0.5:
        return 'Japan'
    elif (other_sales/total) > 0.5:
        return 'Other'
    else:
        return 'Global'
```

```python
df['region_game'] = df.apply(region_game, axis=1)
```

```python
# For the purity of the study, we will construct a schedule for the lack of rating only for games issued since 2000
df[(df['rating'].isna()) & (df['year_of_release'] >=2000)]\
    ['region_game'].value_counts().plot(kind='bar')
plt.title('Number of games without rating by region', fontsize=14)
plt.show()
```

## Check hypothesis

### Medium Custom Ratings Xbox ONE and PC Platforms Same

```python
# to highlight the datamaphraum
df_xone = df.query(
    'platform == "XOne" & user_score <=100 & year_of_release>=2012')
df_pc = df.query('platform == "PC" & user_score <=100 & year_of_release>=2012')
```

#### Distribution normality check

##### Checking when aid chart qq-plot

To correctly select the methodology for testing hypotheses, it is necessary to check the distribution to normal.

Check the distribution to normal with the QQ-Plot graphics

```python
# Build QQ-plot for normal distribution
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
st.probplot(df_xone['user_score'], dist="norm", plot=plt)
plt.subplot(2, 2, 2)
st.probplot(df_pc['user_score'], dist="norm", plot=plt)
plt.show()
```

##### Check with the Criterion Shapiro-Willow

Criterion Shapiro-Willow:

$\begin{equation*}
 \begin{cases}
   H_0 :\text{The average user ratings of the Xbox ONE and PC platforms distributed normally}\\
   H_1 :\text{The average user ratings of the Xbox ONE and PC platforms are not distributed normally.}
 \end{cases}
\end{equation*}$

```python
print("Platform XOne: Shapiro-Wilk normality test, W-statistic: %f, p-value: %f" %
      st.shapiro(df_xone['user_score']))
print("Platform PC: Shapiro-Wilk normality test, W-statistic: %f, p-value: %f" %
      st.shapiro(df_pc['user_score']))
```

**Conclusion**
<div style = "Border: Solid Blue 3px; padding: 20px">
Our distributions differ from normal, so we will apply for testing the hypothesis Mann-Whitney U test. </div>

#### Checking hypothesis

<p>We introduce basic and alternative hypotheses:</p>

  - H_0  Medium Custom Xbox ONE and PC Platform Ratings Same

  - H_1 Medium user ratings of Xbox ONE and PC platforms are different

Significance level: $\alpha = 0.05$

```python
results = st.mannwhitneyu(
    df_xone['user_score'],
    df_pc['user_score'])
print('p-значение: ', results.pvalue)
alpha = .05
if (results.pvalue < alpha):
    print("We reject the zero hypothesis")
else:
    print("It did not work out to reject the zero hypothesis")
```

**Conclusion**
<div style = "Border: Solid Blue 3px; padding: 20px">
On available data, at the level of significance, 5% there is no reason to reject the zero hypothesis in favor of the alternative.
The average user ratings of the Xbox One and PC platforms are identical. </div>
